```python
from lxml import etree
import numpy as np
import pandas as pd
```

```python
def parseHsdbXML(xmlFile):
    
    tree = etree.parse(xmlFile)

    root = tree.getroot()
    
    hsdb = np.array([['name', 'cas', 'bp', 'mp', 'owpc']])
    
    for item in root.findall('DOC'):
        name = item.find('NameOfSubstance').text
        cas = item.find('CASRegistryNumber').text
        bp = item.find('bp').text
        mp = item.find('mp').text
        owpc = item.find('owpc').text
        
        temp = np.array([[name, cas, bp, mp, owpc]])
        
        hsdb = np.concatenate((hsdb, temp))
    
    return hsdb
```

```python
hsdb = parseHsdbXML('data/hsdb/hsdb.xml')
```

```python
import csv
import collections
import re
import io
import json
import xml.etree.ElementTree as ET
import numpy as np
import pandas as pd
```

```python
tree = ET.parse('data/drugbank/full_database.xml')
root = tree.getroot()
```

```python
ns = '{http://www.drugbank.ca}'
inchikey_template = "{ns}calculated-properties/{ns}property[{ns}kind='InChIKey']/{ns}value"
inchi_template = "{ns}calculated-properties/{ns}property[{ns}kind='InChI']/{ns}value"
logP_template = "{ns}experimental-properties/{ns}property[{ns}kind='logP']/{ns}value"
logS_template = "{ns}experimental-properties/{ns}property[{ns}kind='logS']/{ns}value"
pKa_template = "{ns}experimental-properties/{ns}property[{ns}kind='pKa']/{ns}value"
mw_template = "{ns}calculated-properties/{ns}property[{ns}kind='Molecular Weight']/{ns}value"
mp_template = "{ns}experimental-properties/{ns}property[{ns}kind='Melting Point']/{ns}value"
bp_template = "{ns}experimental-properties/{ns}property[{ns}kind='Boiling Point']/{ns}value"

rows = list()
for i, drug in enumerate(root):
    row = collections.OrderedDict()
    assert drug.tag == ns + 'drug'
    row['drugbank_id'] = drug.findtext(ns + "drugbank-id[@primary='true']")
    row['name'] = drug.findtext(ns + "name")
    row['cas'] = drug.findtext(ns + "cas-number")
    row['smiles'] = drug.findtext("{ns}calculated-properties/{ns}property[{ns}kind='SMILES']/{ns}value".format(ns = ns))
    row['groups'] = [group.text for group in
        drug.findall("{ns}groups/{ns}group".format(ns = ns))]
    row['atc_codes'] = [code.get('code') for code in
        drug.findall("{ns}atc-codes/{ns}atc-code".format(ns = ns))]
    row['categories'] = [x.findtext(ns + 'category') for x in
        drug.findall("{ns}categories/{ns}category".format(ns = ns))]
    row['inchi'] = drug.findtext(inchi_template.format(ns = ns))
    row['inchikey'] = drug.findtext(inchikey_template.format(ns = ns))
    row['log_p'] = drug.findtext(logP_template.format(ns = ns))
    row['log_s'] = drug.findtext(logS_template.format(ns = ns))
    row['pka'] = drug.findtext(pKa_template.format(ns = ns))
    row['mw'] = drug.findtext(mw_template.format(ns = ns))
    row['mp'] = drug.findtext(mp_template.format(ns = ns))
    row['bp'] = drug.findtext(bp_template.format(ns = ns))
    
    # Add drug aliases
    aliases = {
        elem.text for elem in 
        drug.findall("{ns}international-brands/{ns}international-brand".format(ns = ns)) +
        drug.findall("{ns}synonyms/{ns}synonym[@language='English']".format(ns = ns)) +
        drug.findall("{ns}international-brands/{ns}international-brand".format(ns = ns)) +
        drug.findall("{ns}products/{ns}product/{ns}name".format(ns = ns))

    }
    aliases.add(row['name'])
    row['aliases'] = sorted(aliases)

    rows.append(row)
```

```python
alias_dict = {row['drugbank_id']: row['aliases'] for row in rows}
with open('./data/aliases.json', 'w') as fp:
    json.dump(alias_dict, fp, indent=2, sort_keys=True)
```

```python
def collapse_list_values(row):
    for key, value in row.items():
        if isinstance(value, list):
            row[key] = '|'.join(value)
    return row

rows = list(map(collapse_list_values, rows))
```

```python
columns = ['drugbank_id','name','cas','smiles','groups','atc_codes','categories','inchikey','inchi','log_p','log_s','pka','mw','mp','bp']
drugbank_df = pd.DataFrame.from_dict(rows)[columns]
drugbank_df.head()
```
